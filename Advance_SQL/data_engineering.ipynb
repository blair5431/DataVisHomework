```python
# Dependencies
import pandas as pd
import numpy as np
import os
```

```python
# importing measurments data 
csvfile1 = "hawaii_measurements.csv"
# Read CSV file into a pandas DataFrame
measurements_df = pd.read_csv(csvfile1, dtype=object)

measurements_df.head()
```

```python
#cleaning the measurment data 

#dropping missing data 
#axis = 0 --> means it is going through the rows
measurements_df = measurements_df.dropna(axis=0)
measurements_df.head()

#reseting the index after dropping missing data 
measurements_df = measurements_df.reset_index(drop=True)
measurements_df.head()
measurements_df["date"]
```

```python
#saving new df
new_csv = "measurements_cleaned.csv"
measurements_df.to_csv(new_csv, index=False)
```

```python
# importing stations data 
csvfile2 = "hawaii_stations.csv"
# Read CSV file into a pandas DataFrame
stations_df = pd.read_csv(csvfile2, dtype=object)

stations_df.head(10)
```

```python
#cleaning the station data 

#dropping missing data 
#axis = 0 --> means it is going through the rows
stations_df = stations_df.dropna(axis=0)
stations_df.head()

#reseting the index after dropping missing data 
stations_df = stations_df.reset_index(drop=True)
stations_df.head()
```

```python
#saving cleaned station data
new_csv2 = "stations_cleaned.csv"
stations_df.to_csv(new_csv2, index=False)
```
